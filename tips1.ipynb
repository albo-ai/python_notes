```python
# 27) Find the first repeated word in a string in Python using Dictionary

# Function to Find the first repeated word in a string 
from collections import Counter 

def firstRepeat(input): 
	
	# first split given string separated by space into words 
	words = input.split(' ') 
	
	# now convert list of words into dictionary 
	dict = Counter(words) 

	# traverse list of words and check which first word has frequency > 1 
	for key in words: 
		if dict[key]>1: 
			print(key)
			return

# Driver program 
if __name__ == "__main__": 
	input = 'Ravi had been saying that he had been there'
	firstRepeat(input) 
```

```python
# 39)  operator overloading
class Thing:
    def __init__(self, value):
        self.__value = value
    def __gt__(self, other):
        return self.__value > other.__value
    def __lt__(self, other):
        return self.__value < other.__value
    def __add__(self, other):
        return self.__value + other.__value
    def __str__(self):
        return "value of class is " + str(self.__value)
    def __repr__(self):
        return "value of class is: " + str(self.__value)



something = Thing(100)
nothing = Thing(1)

del str

print(something > nothing)
print(something < nothing)
print(something + nothing)
print(str(something))
```

```python
# 72) generator example
def fib(num):
    a,b = 0,1
    for i in range(0, num):
        # print "{}: {}".format(i+1, a)  # This will cause a problem in the following loop
        yield "{}: {}".format(i+1, a)  # This will cause a problem in the following loop
        a, b = b, a + b

for item in fib(10):
    print(item)

    
#  why using generators?
```

```python
# 74)  OOP example

class Person(object):
    def __init__(self, name):
        self.name = name

    def reveal_identity(self):
        print("My name is {}".format(self.name))

        
class SuperHero(Person):
    def __init__(self, hero, name):
        super(SuperHero, self).__init__(name)
        self.hero = hero

    def reveal_identity(self):
        super(SuperHero, self).reveal_identity()
        print("My name is {}".format(self.hero))


p = Person('ali')
p.reveal_identity()

ph = SuperHero('yay', 'ali')
ph.reveal_identity()
```

```python
# see also 
# https://www.geeksforgeeks.org/oop-in-python-set-3-inheritance-examples-of-object-issubclass-and-super/

# Another example
class Person(object): 
    # Constructor 
    def __init__(self, name): 
        self.name = name
        
    # To get name 
    def getName(self):
        return self.name

    # To check if this person is employee 
    def isEmployee(self):
        return False
    
        
        
# Inherited or Sub class (Note Person in bracket) 
class Employee(Person):
    def isEmployee(self): 
        return True


# Driver code 
emp = Person("Geek1") # An Object of Person 
print(emp.getName(), emp.isEmployee()) 

emp = Employee("Geek2") # An Object of Employee 
print(emp.getName(), emp.isEmployee()) 
```
